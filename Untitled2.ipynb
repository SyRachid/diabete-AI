```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#chargement des données
data = pd.read_csv('diabetes_prediction_dataset.csv',sep=';')
```

```python
#etape de visualisation et de selection des variables idéales
```

```python
#traitement de la daba en vue de la prediction
```

```python
#traitement de la variable catégorielle genre 
for i in range(100000):
    if data['gender'].loc[i] == 'Female':
        data['gender'].loc[i] = 1
    else:
        data['gender'].loc[i]= 0
```

```python
data['smoking_history'].unique()
```

```python
#traitement de la variable smoking history
for i in range(100000):
    if data['smoking_history'].loc[i] == 'never':
        data['smoking_history'].loc[i] = 0
    if data['smoking_history'].loc[i] == 'No Info':
        data['smoking_history'].loc[i]= 0.5
    if data['smoking_history'].loc[i] == 'current':
        data['smoking_history'].loc[i] =2
    if data['smoking_history'].loc[i] == 'former':
        data['smoking_history'].loc[i] = 1
    if data['smoking_history'].loc[i] == 'ever':
        data['smoking_history'].loc[i] = 0.1
    if data['smoking_history'].loc[i] == 'not current':
        data['smoking_history'].loc[i] = 1.5
    
```

```python
#séparation des données
data_train,data_test = train_test_split(data, test_size=0.2, random_state=42)
X_train = data_train.drop(['diabetes'],axis=1)
X_test = data_test.drop(['diabetes'],axis=1)
y_train = data_train['diabetes']
y_test = data_test['diabetes']
```

```python
#entrainement du modèle Logistic Regression
model_logi = LogisticRegression()
model_logi.fit(X_train,y_train)
```

```python
#prediction du modèle
pred = model_logi.predict(X_test)
```

```python
#analyse du score du modèle 
accuracy = accuracy_score(y_test, pred)
accuracy
```

```python
#exportation du modèle 
import pickle

# Sauvegarder le modèle
with open('model.pkl', 'wb') as f:
    pickle.dump(model_logi, f)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as lr
import numpy as np
```

```python
data=pd.read_csv("sleep.csv",sep=',')
data
```

```python
#pretraitement des donnÃ©es
data = data.drop('Person ID',axis=1)
```

```python
#modification des donnÃ©es de la colonne Sleep Disorder
for i in range(374):
    if data['Sleep Disorder'].iloc[i] == 'None':
        data['Sleep Disorder'].iloc[i] = 0.001
    elif data['Sleep Disorder'].iloc[i] == 'Sleep Apnea':
        data['Sleep Disorder'].iloc[i] = 0.599;
    elif data['Sleep Disorder'].iloc[i] == 'Insomnia':
        data['Sleep Disorder'].iloc[i] = 0.999
```

```python
#modification des donnÃ©es de la colonne Gender
for i in range(374):
    if data['Gender'].iloc[i] == 'Female':
        data['Gender'].iloc[i] = 0.999
   # else :
    #      data['Gender'].iloc[i] == 0.999
   
```
